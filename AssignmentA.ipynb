**Exercise 3** (2 x 12.5 points)
Make yourself familiar with the multinomial distribution

a. Calculate the probability of the counts of outcome “2” if the possible equally probable outcomes are all integer numbers between 1 and 10 and we performed n=1000n=1000 trials.

b. Simulate n independent trials of the multinoulli (categorical distribution) compliant to the specification of (a). Plot the probability in (a) as a function of n independent trials (n=10-1000). Write your conclusions with respect to the behavior of the estimated probability as nn increases.

```python
#Exercise 3A

from numpy import random
import matplotlib.pyplot as plt

px = random.multinomial(1000,[1/10.]*10,size=1) #rolling a 10 sided die 1000 times
print(px)
print(f'Probability for the counts of 2 in 1000 trials - {px[0][1]/1000}')

#Exercise 3B
x = [*range(10,1000+1)]
y=[]
for n in x:
  result = random.multinomial(n,[1/10.]*10,size=1)
  y.append(result[0][1]/n)  


fig = plt.figure(figsize=(10,7))
ax=fig.add_subplot(111)
ax.plot(x,y)
ax.set_xlabel("trials")
ax.set_ylabel('probability')
ax.set_title('probability for counts of outcome 2 over 10 - 1000 trials')
```

conclusion from the above graph: As n increases, probability for counts of 2 gets more or less the consistent with previous trials
